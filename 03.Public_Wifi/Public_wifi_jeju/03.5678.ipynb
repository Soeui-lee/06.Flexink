#### 6.1. 실제입도객수와 비교

**공공 와이파이는 관광객 뿐 아니라 도민, 관광 외 목적 입도객도 사용**
> 
실제 관광객이라 하더라도 **데이터, 도시락 등** 공공 와이파이 외 선택지가 있음 
>
그럼에도 추이가 비슷하다면 관광객에 빗댈 수 있다고 가정하고 사용

```python
jeju_travel = pd.read_csv('제주관광통계_20,21_5678.csv')    # 제주특별자치도관광협회 제주관광통계 
travel20 = jeju_travel[jeju_travel['Year ']==2020]
travel21 = jeju_travel[jeju_travel['Year ']==2021]
```

```python
travel20['ap20_acc'] = np.nan
travel21['ap21_acc'] = np.nan

travel20['ap20_acc'][0] = len(ap20[ap20['pt_month'] == 5]['user_id'].unique())
travel20['ap20_acc'][1] = len(ap20[ap20['pt_month'] == 6]['user_id'].unique())
travel20['ap20_acc'][2] = len(ap20[ap20['pt_month'] == 7]['user_id'].unique())
travel20['ap20_acc'][3] = len(ap20[ap20['pt_month'] == 8]['user_id'].unique())

travel21['ap21_acc'][4] = len(ap21[ap21['pt_month'] == 5]['user_id'].unique())
travel21['ap21_acc'][5] = len(ap21[ap21['pt_month'] == 6]['user_id'].unique())
travel21['ap21_acc'][6] = len(ap21[ap21['pt_month'] == 7]['user_id'].unique())
travel21['ap21_acc'][7] = len(ap21[ap21['pt_month'] == 8]['user_id'].unique())
```

### 7. 2021년 지역별 분석

- 제주 동부 지역별( +서귀포 성산읍) 데이터프레임화

```python
# 우도면 
Udo_21 = ap21[ap21['ap_address_detail'].str.contains('우도', na=False)]

# 조천읍
Jocheon_21 = ap21[ap21['ap_address_detail'].str.contains('조천', na=False)]

# 구좌읍
Gujwa_21 = ap21[ap21['ap_address_detail'].str.contains('구좌', na=False)]

# 서귀포 성산읍
Seongsan_21 = ap21[ap21['ap_address_detail'].str.contains('성산', na=False)]

# 구제주
Gujeju_21 = ap21[ap21['ap_address_detail'].str.contains('일도|이도|삼도|용담|건입|도두|화북|삼양|아라|봉개', na=False)]

# 알 수 없음(결측처리)
X_21 = ap21[ap21['ap_address_detail'].str.contains('알 수 없음', na=False)]
```
